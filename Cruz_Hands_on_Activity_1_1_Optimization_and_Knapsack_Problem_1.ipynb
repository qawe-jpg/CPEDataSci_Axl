```python
class Food(object):
    def __init__(self, name, value, calories):
        # Make the variables private
        self._name = name
        self._value = value
        self._calories = calories

    # Returns the value of the food
    def getValue(self):
        return self._value

    # Returns the cost (calories) of the food
    def getCost(self):
        return self._calories

    # Calculates and returns the density (value / cost)
    def density(self):
        return self.getValue() / self.getCost()

    # Returns a string representation of the food
    def __str__(self):
        return f"{self._name}: <{self._value}, {self._calories}>"
```

4. Create a method greedy to return total value and cost of added food based on the desired maximum cost

```python
def greedy(items, maxCost, keyFunction):
    
    itemsCopy = sorted(items, key=keyFunction, reverse=True)

    result = []
    totalValue, totalCost = 0.0, 0.0

    for item in itemsCopy:
        if totalCost + item.getCost() <= maxCost:
            result.append(item)
            totalCost += item.getCost()
            totalValue += item.getValue()

    return result, totalValue, totalCost
```

5. Create a testGreedy method to test the greedy method

```python
def testGreedy(items, constraint, keyFunction):
    taken, totalValue, totalCost = greedy(items, constraint, keyFunction)

    print('Total value of items taken =', totalValue)
    print('Total cost of items taken =', totalCost)

    for item in taken:
        print('   ', item)
```

9. Create method to use  Bruteforce algorithm instead of greedy algorithm

```python
def maxVal(toConsider, avail):
    """
    Assumes toConsider is a list of Food items
    avail is the maximum allowed cost (calories)
    Returns a tuple:
        (total value, tuple of selected items)
    """
    if toConsider == [] or avail == 0:
        return (0, ())

    # If item is too heavy, skip it
    if toConsider[0].getCost() > avail:
        return maxVal(toConsider[1:], avail)

    nextItem = toConsider[0]

    # Explore left branch (take item)
    withVal, withToTake = maxVal(
        toConsider[1:], avail - nextItem.getCost()
    )
    withVal += nextItem.getValue()

    # Explore right branch (skip item)
    withoutVal, withoutToTake = maxVal(
        toConsider[1:], avail
    )

    # Choose better branch
    if withVal > withoutVal:
        return (withVal, withToTake + (nextItem,))
    else:
        return (withoutVal, withoutToTake)
```

```python
def testMaxVal(foods, maxUnits, printItems=True):
    print('Use search tree to allocate', maxUnits, 'calories')

    val, taken = maxVal(foods, maxUnits)

    print('Total value of foods taken =', val)
    if printItems:
        for item in taken:
            print('   ', item)
```

#### Supplementary Activity:

* Choose a real-world problem that solves knapsacks problem
* Use the greedy and brute force algorithm to solve knapsacks problem


```python
class Package:
    def __init__(self, name, value, weight):
        self.name = name
        self.value = value
        self.weight = weight

    def density(self):
        return self.value / self.weight

    def __str__(self):
        return f"{self.name}: <value={self.value}, weight={self.weight}>"

def buildPackageList(names, values, weights):
    return [Package(names[i], values[i], weights[i]) for i in range(len(names))]

def greedy(items, maxWeight, keyFunc):
    items = sorted(items, key=keyFunc, reverse=True)
    taken, totalValue, totalWeight = [], 0, 0
    for item in items:
        if totalWeight + item.weight <= maxWeight:
            taken.append(item)
            totalValue += item.value
            totalWeight += item.weight
    return taken, totalValue

def maxVal(items, avail):
    if not items or avail == 0:
        return 0, ()
    if items[0].weight > avail:
        return maxVal(items[1:], avail)
    else:
        withVal, withItems = maxVal(items[1:], avail - items[0].weight)
        withVal += items[0].value
        withoutVal, withoutItems = maxVal(items[1:], avail)
        if withVal > withoutVal:
            return withVal, withItems + (items[0],)
        else:
            return withoutVal, withoutItems

names = ['Laptop', 'TV', 'Fridge', 'Chair', 'Table', 'Microwave', 'Phone', 'Bicycle', 'Printer']
values = [1000, 800, 1200, 300, 400, 250, 700, 600, 350]
weights = [5, 20, 80, 10, 15, 8, 3, 25, 7]

packages = buildPackageList(names, values, weights)
maxCapacity = 100

taken, val = greedy(packages, maxCapacity, Package.density)
print("Greedy selection:")
for p in taken:
    print(" ", p)
print("Total value:", val)

val, taken = maxVal(packages, maxCapacity)
print("\nBrute-force optimal selection:")
for p in taken:
    print(" ", p)
print("Total value:", val)
```
