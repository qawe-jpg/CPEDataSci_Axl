```python
#plotting with pandas
#set up
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

fb = pd.read_csv(
    'fb_stock_prices_2018.csv',
    index_col='date',
    parse_dates=True
)

quakes = pd.read_csv('earthquakes-1.csv')
```

```python
#evolution overtime
fb.plot(
    kind='line',
    y='open',
    figsize=(10, 5),
    style='b-',
    legend=False,
    title='Evolution of Facebook Open Price'
)
```

```python
fb.plot(
kind='line',
y='open',
figsize=(10, 5),
color='blue',
linestyle='solid',
legend=False,
title='Evolution of Facebook Open Price'
)
```

```python
#creating subplots
fb.plot(
kind='line',
subplots=True,
layout=(3,2),
figsize=(15,10),
title='Facebook Stock 2018'
)
```

```python
#scatter plots
fb.assign(
max_abs_change=fb.high - fb.low
).plot(
kind='scatter', x='volume', y='max_abs_change',
title='Facebook Daily High - Low vs. Volume Traded'
)
```

```python
fb.assign(
max_abs_change=fb.high - fb.low
).plot(
kind='scatter', x='volume', y='max_abs_change',
title='Facebook Daily High - Low vs. log(Volume Traded)',
logx=True
)
```

```python
#Adding Transparency to Plots with alpha
fb.assign(
max_abs_change=fb.high - fb.low
).plot(
kind='scatter', x='volume', y='max_abs_change',
title='Facebook Daily High - Low vs. log(Volume Traded)',
logx=True, alpha=0.25
)
```

```python
#hexbins
fb.assign(
log_volume=np.log(fb.volume),
max_abs_change=fb.high - fb.low
).plot(
kind='hexbin',
x='log_volume',
y='max_abs_change',
title='Facebook Daily High - Low vs. log(Volume Traded)',
colormap='gray_r',
gridsize=20,
sharex=False # we have to pass this to see the x-axis due to a bug in this version of pandas
)
```

```python
#Visualizing Correlations with Heatmaps
fig, ax = plt.subplots(figsize=(20, 10))

# Compute correlation matrix with added features
fb_corr = fb.assign(
    log_volume=np.log(fb.volume),
    max_abs_change=fb.high - fb.low
).corr()

# Display matrix
im = ax.matshow(fb_corr, cmap='seismic', vmin=-1, vmax=1)
fig.colorbar(im)

# Set tick positions and labels
labels = [col.lower() for col in fb_corr.columns]
ax.set_xticks(range(len(labels)))
ax.set_yticks(range(len(labels)))
ax.set_xticklabels(labels, rotation=45, ha='left')
ax.set_yticklabels(labels)

plt.show()
```

```python
#Stacked bar chart
pivot = quakes.assign(
mag_bin=lambda x: np.floor(x.mag)
).pivot_table(
index='mag_bin', columns='magType', values='mag', aggfunc='count'
)
pivot.plot.bar(
stacked=True, rot=0,
title='Earthquakes by integer magnitude and magType'
)
plt.ylabel('earthquakes') # label the axes (discussed in chapter 6)
```

```python
#Normalized stacked bars
normalized_pivot = pivot.fillna(0).apply(lambda x: x/x.sum(), axis=1)
ax = normalized_pivot.plot.bar(
stacked=True, rot=0, figsize=(10, 5),
title='Percentage of earthquakes by integer magnitude for each magType'
)
ax.legend(bbox_to_anchor=(1, 0.8)) # move legend to the right of the plot
plt.ylabel('percentage') # label the axes (discussed in chapter 6)
```

```python
#supplementary activities
#1.Rolling 20-day minimum of Facebook closing price
import pandas as pd
import matplotlib.pyplot as plt

fb = pd.read_csv('fb_stock_prices_2018.csv', index_col='date', parse_dates=True)


fb['close'].rolling(window=20).min().plot(
    figsize=(10,5),
    title='20-Day Rolling Minimum of Facebook Closing Price',
    color='blue'
)
plt.ylabel('Price ($)')
plt.show()
```

```python
#2.Histogram and KDE of daily change (close - open)
fb['daily_change'] = fb['close'] - fb['open']


fb['daily_change'].plot.hist(
    bins=30,
    density=True,
    alpha=0.5,
    color='skyblue',
    figsize=(10,5),
    title='Histogram and KDE of Daily Change in Facebook Price'
)
fb['daily_change'].plot.kde(color='red')
plt.xlabel('Daily Change ($)')
plt.show()
```

```python
#3.Box plots of earthquake magnitudes by magType in Indonesia
quakes = pd.read_csv('earthquakes-1.csv')


indonesia_quakes = quakes[quakes['place'].str.contains('Indonesia', case=False, na=False)]


indonesia_quakes.boxplot(column='mag', by='magType', figsize=(8,5))
plt.title('Earthquake Magnitudes by magType in Indonesia')
plt.suptitle('')
plt.ylabel('Magnitude')
plt.show()
```

```python
#4.Weekly line plot of high-low difference
weekly_diff = fb['high'].resample('W').max() - fb['low'].resample('W').min()

weekly_diff.plot(
    figsize=(10,5),
    title='Weekly High-Low Difference of Facebook Stock'
)
plt.ylabel('Price Difference ($)')
plt.show()
```

```python
#5.Subplots showing after-hours effects
fig, axes = plt.subplots(1, 2, figsize=(16,5))


after_hours = fb['open'] - fb['close'].shift(1)
after_hours.plot(ax=axes[0], title='Daily After-Hours Effect', color='purple')
axes[0].set_ylabel('Price Difference ($)')


monthly_net = after_hours.resample('M').sum()


colors = ['green' if x >=0 else 'red' for x in monthly_net]
monthly_net.plot.bar(ax=axes[1], color=colors, title='Monthly Net After-Hours Effect')
axes[1].set_ylabel('Net Change ($)')


axes[1].set_xticklabels([d.strftime('%b') for d in monthly_net.index])

plt.tight_layout()
plt.show()
```
