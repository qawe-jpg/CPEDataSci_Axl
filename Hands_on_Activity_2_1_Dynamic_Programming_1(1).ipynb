1. Create a code that demonstrate how to use recursion method to solve problem

```python
def factorial(n):
    if n == 1:
        return 1
    return n * factorial(n-1)

def sum_of_factorials(n):
    if n == 1:
        return 1
    return factorial(n) + sum_of_factorials(n-1)

print(sum_of_factorials(4))
```

2. Create a program codes that demonstrate how to use dynamic programming to solve the same problem

```python
def sum_of_factorials(n):
    dp = [0] * (n + 1)
    dp[1] = 1

    for i in range(2, n + 1):
        dp[i] = dp[i-1] * i

    return sum(dp[1:])

print(sum_of_factorials(4))
```

##### Question:
Explain the difference of using the recursion from dynamic programming using the given sample codes to solve the same problem


Recursion solves the problem by repeatedly calling functions, which causes the same factorial values to be recomputed many times.
Dynamic programming stores previously computed factorial values in a list, so each value is calculated only once.
As a result, dynamic programming is faster and more efficient than recursion for the same proble





3. Create a sample program codes to simulate bottom-up dynamic programming

```python
def fib(n):
    dp = [0, 1]
    for i in range(2, n + 1):
        dp.append(dp[i-1] + dp[i-2])
    return dp[n]

print(fib(6))
```

4. Create a sample program codes that simulate tops-down dynamic programming

```python
def fib(n, memo={}):
    if n in memo:
        return memo[n]
    if n <= 1:
        return n

    memo[n] = fib(n-1, memo) + fib(n-2, memo)
    return memo[n]

print(fib(6))
```

#### Question:
 Explain the difference between bottom-up from top-down dynamic programming using the given sample codes

Bottom up dynamic programming solves the problem iteratively from the smallest subproblems, building up to the final solution unlike
top down dynamic programming uses recursion and stores results in memory to avoid recalculating the same subproblems.
Bottom up is usually more memory and time efficient, while top down is easier to write and understand for recursive problems.




0/1 Knapsack Problem

* Analyze three different techniques to solve knapsacks problem
1. Recursion
2. Dynamic Programming
3. Memoization

```python
#sample code for knapsack problem using recursion
def rec_knapSack(w, wt, val, n):

  #base case
  #defined as nth item is empty;
  #or the capacity w is 0
  if n == 0 or w == 0:
    return 0

  #if weight of the nth item is more than
  #the capacity W, then this item cannot be included
  #as part of the optimal solution
  if(wt[n-1] > w):
    return rec_knapSack(w, wt, val, n-1)

  #return the maximum of the two cases:
  # (1) include the nth item
  # (2) don't include the nth item
  else:
    return max(
        val[n-1] + rec_knapSack(
            w-wt[n-1], wt, val, n-1),
            rec_knapSack(w, wt, val, n-1)
    )
```

```python
#To test:
val = [60, 100, 120] #values for the items
wt = [10, 20, 30] #weight of the items
w = 50 #knapsack weight capacity
n = len(val) #number of items

rec_knapSack(w, wt, val, n)
```

```python
#Dynamic Programming for the Knapsack Problem
def DP_knapSack(w, wt, val, n):
  #create the table
  table = [[0 for x in range(w+1)] for x in range (n+1)]

  #populate the table in a bottom-up approach
  for i in range(n+1):
    for w in range(w+1):
      if i == 0 or w == 0:
        table[i][w] = 0
      elif wt[i-1] <= w:
        table[i][w] = max(val[i-1] + table[i-1][w-wt[i-1]],
                          table[i-1][w])
  return table[n][w]
```

```python
#To test:
val = [60, 100, 120]
wt = [10, 20, 30]
w = 50
n = len(val)

DP_knapSack(w, wt, val, n)
```

```python
#Sample for top-down DP approach (memoization)
#initialize the list of items
val = [60, 100, 120]
wt = [10, 20, 30]
w = 50
n = len(val)

#initialize the container for the values that have to be stored
#values are initialized to -1
calc =[[-1 for i in range(w+1)] for j in range(n+1)]


def mem_knapSack(wt, val, w, n):
  #base conditions
  if n == 0 or w == 0:
    return 0
  if calc[n][w] != -1:
    return calc[n][w]

  #compute for the other cases
  if wt[n-1] <= w:
    calc[n][w] = max(val[n-1] + mem_knapSack(wt, val, w-wt[n-1], n-1),
                     mem_knapSack(wt, val, w, n-1))
    return calc[n][w]
  elif wt[n-1] > w:
    calc[n][w] = mem_knapSack(wt, val, w, n-1)
    return calc[n][w]

mem_knapSack(wt, val, w, n)
```

**Code Analysis**




In the first code using recursion, the program repeatedly calls itself for every item combination, checking whether to include or exclude each item. This matches the recursive function rec_knapSack, where each decision leads to two more calls, which can be very slow for many items. The second code, using bottom-up dynamic programming, builds a table iteratively with DP_knapSack, storing the best value for each possible weight and number of items, so no calculation is repeated. The third code, using top-down dynamic programming or memorization with mem_knapSack, keeps the recursive structure but stores results in the calc table, avoiding repeated calculations of the same sub problems. Both dynamic programming approaches are more efficient than plain recursion while still guaranteeing the maximum value.

## Supplementary Problem (HOA 2.1 Submission):

```python
#type your code here for recursion programming solution
def climb_stairs_rec(n, broken):
    if n == 0:
        return 1
    if n < 0 or n in broken:
        return 0
    return climb_stairs_rec(n-1, broken) + climb_stairs_rec(n-2, broken) + climb_stairs_rec(n-3, broken)

n = 7
broken = [3, 5]
print(climb_stairs_rec(n, broken))
```

```python
#type your code here for dynamic programming solution
def climb_stairs_dp(n, broken):
    dp = [1] + [0]*n
    for i in range(1, n+1):
        if i in broken:
            dp[i] = 0
        else:
            dp[i] = dp[i-1]
            if i-2 >= 0:
                dp[i] += dp[i-2]
            if i-3 >= 0:
                dp[i] += dp[i-3]
    return dp[n]

print(climb_stairs_dp(n, broken))
```
