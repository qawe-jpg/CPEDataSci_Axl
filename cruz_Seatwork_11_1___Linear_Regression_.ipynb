**Part 1**

```python
#part 1
import matplotlib.pyplot as plt
import numpy as np


b = 1
w1 = 3
x1 = 0


y = b + w1 * x1
print("Prediction:", y)


x_vals = np.linspace(-5, 5, 100)
y_vals = b + w1 * x_vals

plt.plot(x_vals, y_vals)
plt.scatter(x1, y)
plt.title("Linear Regression (1 feature)")
plt.xlabel("x1")
plt.ylabel("y")
plt.grid()

plt.show()
```
